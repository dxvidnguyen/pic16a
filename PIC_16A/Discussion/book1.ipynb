```python
#prepping data
subset = penguins[['Species', 'Flipper Length (mm)', 'Body Mass (g)', 'Sex']]
subset = subset.dropna() #drop nan values
subset.info()
```

```python
subset.head()
```

```python
x_slice = subset[["Flipper Length (mm)", "Body Mass (g)", "Sex"]]
y_slice = subset[['Species']]

print(x_slice.head())
print(y_slice.head())
```

```python
le = preprocessing.LabelEncoder()
le.fit_transform(x_slice["Sex"])
'''
0 is a period 
1 is female
2 is male
'''
```

```python
x_slice.head()
```

```python
x_slice['Sex'] = le.fit_transform(x_slice["Sex"])
x_slice.head()
```

```python
le.fit_transform(y_slice['Species'])
'''
0 = adelie 
1 = chinstap
2 = gentoo
'''

y_slice['Species'] = le.fit_transform(y_slice['Species'])
y_slice.head()
```

```python
#splitting up the data into test and training sets
x_train, x_test, y_train, y_test = train_test_split(x_slice, y_slice, test_size = 0.2)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score
```

```python
LR = LogisticRegression()
cross_val_score(LR, x_train, y_train, cv = 5).mean()
```
